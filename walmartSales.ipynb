# Estimate the Unit Sales of Walmart
Use hierarchical sales data from Walmart, the world’s largest company by revenue, to forecast daily sales for the next 28 days using machine learning. The data, covers stores in three US States (California, Texas, and Wisconsin) and includes item level, department, product categories, and store details. In addition, it has explanatory variables such as price, promotions, day of the week, and special events. Together, this robust dataset can be used to improve forecasting accuracy. 

The methods used can be applied in various business areas, such as setting up appropriate inventory or service levels. Through its business support and training, the MOFC will help distribute the tools and knowledge so others can achieve more accurate and better calibrated forecasts, reduce waste and be able to appreciate uncertainty and its risk implications. The original dataset and information can be found on Kaggle:
https://www.kaggle.com/competitions/m5-forecasting-accuracy

## Datasets
There are three main datasets which are calendar, sell_prices, and train.

```python
import pandas as pd
import numpy as np
calendar=r'/kaggle/input/m5-forecasting-accuracy/calendar.csv'
sell_prices='/kaggle/input/m5-forecasting-accuracy/sell_prices.csv'
train=r'/kaggle/input/m5-forecasting-accuracy/sales_train_validation.csv'
```

## Train Data

```python
train=pd.read_csv(train)
train.head()
```

## Sell Prices

```python
sell_prices=pd.read_csv(sell_prices)
sell_prices.head()
```

## Calender

```python
calendar=pd.read_csv(calendar)
calendar.head()
```

## New Price Merged Dataset

```python
priceDF=calendar[['date', 'wm_yr_wk']].merge(sell_prices, on=['wm_yr_wk'], how='left')
priceDF['id']=priceDF['item_id']+'_'+priceDF['store_id']+'_validation'
priceDF.head()
```

```python
#JOIN NEW priceDF id column with trainDF.
finalDF=train.merge(priceDF, on=['id'], how='left')
finalDF.head()
```
